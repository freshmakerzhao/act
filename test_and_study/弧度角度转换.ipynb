```python
import math
import numpy as np
```

```python

# ==========================================
# 方法一：使用内置 math 库（适合单个数值转换）
# ==========================================
rad_val = 1.45
deg_val = math.degrees(rad_val)
print(f"[单个转换] {rad_val} 弧度 = {deg_val:.2f} 度")

deg_input = 90
rad_output = math.radians(deg_input)
print(f"[单个转换] {deg_input} 度 = {rad_output:.4f} 弧度\n")
```

```python

# ==========================================
# 方法二：使用 numpy（适合你直接处理 qpos 字符串或数组）
# ==========================================
qpos_string = "0 -1.54 1.45 -1.4 -1.56 0"

# 1. 将字符串拆分并转换为 float 类型的 numpy 数组
qpos_rad_array = np.array([float(x) for x in qpos_string.split()])

# 2. 批量将弧度转换为角度 (np.degrees 或 np.rad2deg 都可以)
qpos_deg_array = np.degrees(qpos_rad_array)

print("机械臂前 6 个关节的弧度:", qpos_rad_array)
print("转换后的直观角度 (度):", np.round(qpos_deg_array, 2))

# 反向转换回去：角度转弧度
# qpos_rad_back = np.radians(qpos_deg_array)
```
